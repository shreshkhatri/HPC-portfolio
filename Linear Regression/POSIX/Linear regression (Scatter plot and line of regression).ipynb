```python
#loading of numpy and pandas library to work with arrays and dataframes read from csv files
import pandas as pd

#loading of pyplot lib for scatter plot and line of regression
import matplotlib.pyplot as pplot
```

```python
#reading of datasets
dataset=pd.read_csv('a/outputs/lr00_results.csv')

#printing first 5 data from dataset 
dataset.head()

#functions for calculating x and y co-oridnates for given value of slope and intercept
def plot_line(slope, intercept,X,Y):
    for i in range(100):
        x=i
        X.append(x)
        y=slope*x+intercept
        Y.append(y)
```

```python
#########################################################################33
#   SCATTER PLOT 
###################################################################33333

#reading of x co-ordinates
x_coordinates=dataset['x']

#reading of y co-ordinates
y_coordinates=dataset['y']


#setting Graph size plotted by pyplot
pplot.rcParams['figure.figsize']=(13,8)
pplot.scatter(x_coordinates,y_coordinates,s=1,label='single data point')
pplot.xlabel('X co-ordinate')
pplot.ylabel('Y co-ordinate')
pplot.title('Scatter plot of datasets')


pplot.legend()
pplot.show()

```

```python
##########################################################################
#   3 LINES WITH 3 OPTIMUM GUESSES                                       #
##########################################################################

#lists for holding x co-ordinates
X1=[] 
X2=[] 
X3=[]

#lists for holding y co-ordinates
Y1=[] 
Y2=[] 
Y3=[]


plot_line(1.6,20,X1,Y1)
plot_line(1.7,18,X2,Y2)
plot_line(1.6,17,X3,Y3)

#setting Graph size plotted by pyplot
pplot.rcParams['figure.figsize']=(13,8)
pplot.scatter(x_coordinates,y_coordinates,s=1,label='single data point')
pplot.xlabel('X co-ordinate')
pplot.ylabel('Y co-ordinate')
pplot.title('Scatter plot of datasets and regression lines')
#drawing lines
pplot.plot(X1,Y1,label='m = 1.6, c = 20, error =10.576',color='r')
pplot.plot(X2,Y2,label='m = 1.7, c = 18 error =12.7885',color='darkgrey')
pplot.plot(X2,Y3,label='m = 1.6, c = 17, error =10.099',color='limegreen')


pplot.legend()
pplot.show()



```

```python
  ##########################################################################################
 #   OVERLAYING SOLUTION PRESENTED BY SEQUENTIAL VERSION OF THE lINEAR REGRESSION PRGRAM  #
##########################################################################################

##########################################################################
#   3 LINES WITH 3 OPTIMUM GUESSES                                       #
##########################################################################

#lists for holding x co-ordinates
X_POS=[] 

#lists for holding y co-ordinates
Y_POS=[] 

plot_line(1.52,20.88,X_POS,Y_POS)

#setting Graph size plotted by pyplot
pplot.rcParams['figure.figsize']=(13,8)
pplot.scatter(x_coordinates,y_coordinates,s=1,label='single data point')
pplot.xlabel('X co-ordinate')
pplot.ylabel('Y co-ordinate')
pplot.title('Scatter plot of datasets and the regression line')
#drawing lines
pplot.plot(X_POS,Y_POS,label='m = 1.52, c = 20.88, error = 9.738',color='limegreen')


pplot.legend()
pplot.show()



```
