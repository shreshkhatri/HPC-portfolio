```python
import matplotlib.pyplot as pyp
```

```python
  #######################################################################
 #   PASSWORD CRACKING : POSIX VS SEQUENTIAL VS CUDA                   #
#######################################################################

X1=['Sequential\n']
X2=['POSIX\n'] 
X3=['CUDA\n']

Y1=[506.87]
Y2=[260.53]
Y3=[0.074]

pyp.bar(X1,Y1,label='Sequential')
pyp.bar(X2,Y2,label='POSIX')
pyp.bar(X3,Y3,label='CUDA')
pyp.xlabel('Parallel computing Technologies')
pyp.ylabel('Mean execution time (in seconds )')
pyp.title('Mean execution time for differnt Parallel computing Technology\nFor Password Cracking Program\n')
pyp.legend()
pyp.show()
```
